```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.sin(x) + x/10

x = np.linspace(-6, 6, 100)
plt.plot(x, f(x))
plt.grid(True)
plt.xlabel('beta')
plt.ylabel('R')
plt.show()
```

```python
def df(x):
    return np.cos(x) + 1/10

plt.plot(x, df(x))
plt.grid(True)
plt.xlabel('beta')
plt.ylabel("R'")
plt.show()
```

```python
def grad_desc(f, df, x0, eta=0.1, toll=1e-6, max_iter=10000):
    x_old = x0
    x_list = [x0]
    for _ in range(max_iter):
        x_new = x_old - eta*df(x_old)
        if np.abs(x_new - x_old) < toll:
            break
        x_old = x_new
        x_list.append(x_old)
    return x_list

x_all = grad_desc(f, df, x0=2.3)

x_final = x_all.pop()

plt.plot(x, f(x), label="R")
plt.scatter(x_all, f(np.array(x_all)), color='g', label="steps")
plt.scatter(x_final, f(x_final), color="r", label="last step")
plt.grid(True)
plt.xlabel('beta')
plt.ylabel('R')
plt.legend()
plt.show()
```

```python
x_all = grad_desc(f, df, x0=1.4)

x_final = x_all.pop()

plt.plot(x, f(x), label="R")
plt.scatter(x_all, f(np.array(x_all)), color='g', label="steps")
plt.scatter(x_final, f(x_final), color="r", label="last step")
plt.grid(True)
plt.xlabel('beta')
plt.ylabel('R')
plt.legend()
plt.show()
```
